```python
# Mengimport Library
import cv2
import matplotlib.pyplot as plt
```

```python
# Membaca Gambar
img = cv2.imread('pemandangan.jpeg')
```

```python
# Konversi Gambar ke RGB
RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Mengubah gambar ke Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Mengubah gambar ke Biner
ret, biner_img = cv2.threshold(gray_img,128, 255, cv2.THRESH_BINARY)
```

```python
# Menampilkan Gambar
titles = ['Original Image', 'RGB', 'Gray Image', 'Biner Image']
images = [img, RGB, gray_img, biner_img,]

fig, axes = plt.subplots(1, 4, figsize=(12, 8))
fig.subplots_adjust(wspace=0.5, hspace=0.5)

for i in range(4):
  plt.subplot(1, 4, i + 1)
  plt.imshow(images[i], 'gray', vmin=0, vmax=255)
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])

plt.show()
```
